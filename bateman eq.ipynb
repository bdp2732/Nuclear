```
import numpy as np
import matplotlib.pyplot as plt
```

```
dt=0.000001
t = np.arange(0,2,dt)
I=6 # 첫 번째 붕괴상수
J=6 # 두 번째 붕괴상수
K=0 # 세 번째 붕괴상수
dN=lambda M,I,T:-I*M*T #radiocative decay
n0=10000  # 원소의 개수
n1=[n0]
```

```
for i in t:
  n1.append(n1[-1]+dN(n1[-1],I,dt))
n1=np.array(n1)
n1=np.around(n1)
R=list(np.array(n1)-n0)

X=-R[1]
X=int(X)
n2=[X]
```

```
for i in range(len(t)):
  n2.append(n2[-1]-dN(n1[i],I,dt)+dN(n2[-1],J,dt))

n2.insert(0,0)
A=n2[1:]
n2=np.array(n2)
n2=np.around(n2)
A=np.array(A)

T=n1[2]-n1[1]
V=n1[1]-n1[0]
Y=T+V+n2[2]
Y=-int(Y)
n3=[Y]
```

```
for i in range(len(t)):
  n3.append(n3[-1]-dN(A[i],J,dt)+dN(n3[-1],K,dt))

n3.insert(0,0)
n3.insert(1,0)
n3=np.array(n3)
n3=np.around(n3)
```

```
plt.plot(t,n1[:-1], label='Na')
plt.plot(t,n2[:-2], label='Nb')
plt.plot(t,n3[:-3], label='Nc')
plt.xlabel("Time")
plt.ylabel('Relative number of nuclei')
plt.legend()
plt.show()
```
